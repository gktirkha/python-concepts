I am going to fetch email by subject and going to fetch email sent in [sending_email.ipynb](https://github.com/gktirkha/python-concepts/blob/master/email/sending_email.ipynb)

```python
import imaplib
import getpass
```

imap protocol is used to read emails, to use python with gmail imap we need to get gmail imap server<br/>
gmail imap server is imap.gmail.com

```python
imaplib_object = imaplib.IMAP4_SSL('imap.gmail.com')
```

getting email and password 

> Use app password which can be generated by visiting this [link](https://myaccount.google.com/apppasswords)

Getting Email And Password

```python
email = getpass.getpass("Enter Email")
password = password = getpass.getpass("Enter Password")
```

Logging in

```python
imaplib_object.login(email,password)
```

Getting List Of All Mail Box

```python
imaplib_object.list()
```

Selecting inbox

```python
imaplib_object.select('inbox')
```

Searching For subject "Test Subject", Data will contain a lis of email_id (not the sender email account, but index id of email in your mail box) that matches the query

```python
typ,data = imaplib_object.search(None,'SUBJECT "Test Subject"')
```

getting email_id of first result

```python
email_id = data[0]
```

Fetching the email, email_data will contain all the email related information, and result will be OK if fetching was successful

```python
result,email_data = imaplib_object.fetch(email_id,'(RFC822)')
```
